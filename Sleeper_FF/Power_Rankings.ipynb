```python
def two_step_dominance(X):
    '''Returns result of two step dominance formula'''
    X_sq = np.matmul(X,X)
    matrix = np.add(X_sq,X)
    result = [sum(x) for x in matrix]
    return result


def power_points(dominance, teams, week):
    '''Returns list of power points'''
    power_points = []
    for i, team in zip(dominance, teams):
        avg_score = sum(team.scores[:week]) / week
        avg_mov = sum(team.mov[:week]) / week

        power = '{0:.2f}'.format((int(i)*0.8) + (int(avg_score)*0.15) +
                                 (int(avg_mov)*0.05))
        power_points.append(power)
    power_tup = [(i, j) for (i, j) in zip(power_points, teams)]
    return sorted(power_tup, key=lambda tup: float(tup[0]), reverse=True)
```

```python
win_df = pd.DataFrame(np.zeros(shape=(10,10)))
win_df.columns = Freeman.keys()
win_df.index = Freeman.keys()
win_mov = win_df.copy()
for team in Freeman_sched:
    ## loop through team
    opponents = Freeman_sched[team]
    ## loop through opponents for team 
    for week,opp in enumerate(opponents):
        #print(team + " vs. " + opp + " Week " + str(week))
        opp_score = Freeman[opp][week]
        team_score = Freeman[team][week]
        if team_score > opp_score:
            win_df[team][opp] += 1
            win_mov[team][opp] += team_score - opp_score 
```
